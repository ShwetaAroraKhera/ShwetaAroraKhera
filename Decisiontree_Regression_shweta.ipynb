```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/age_height.csv')
data
```

```python
x_age = data.iloc[:,0:1].values
print(type(x_age), x_age.shape)
```

```python
y_height = data.iloc[:,-1].values
print(type(y_height), y_height.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x_age, y_height)
```

```python
plt.scatter(x_age, y_height)
plt.plot(x_age, regressor.predict(x_age), color='red')
plt.show()
```

```python
x_random = np.random.randint(10,80, size= (70,1))
print(type(x_random), x_random.shape)
```

```python
y_random_pred = regressor.predict(x_random)
```

```python
plt.scatter(x_age, y_height)
plt.plot(x_random, y_random_pred, color = 'green')
plt.show()
```

**Split the data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_age, y_height, test_size=0.30, random_state=0)
```

**Fit data into model**

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)  ## predict the height for test age data using the training data
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_test, y_test, color = 'green')
plt.plot(X_test, y_pred, color = 'blue')
plt.show()
```

```python
x_random_test = np.random.randint(10,80, size= (70,1))
print(type(x_random_test), x_random_test.shape)
```

```python
y_random_test_pred = regressor.predict(x_random_test) ## use the split training data to predict the height for the random data
```

```python
plt.scatter(X_train, y_train)
plt.plot(x_random, y_random_pred, color = 'green')
plt.show
```
